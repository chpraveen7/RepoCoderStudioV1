## 4. Inspect predictions


```python
import glob, json
pred_file = sorted(glob.glob(summary['run_dir'] + '/predictions__*.jsonl'))[0]
for line in open(pred_file):
    r = json.loads(line)
    print('CODE:', r['code'].splitlines()[0])
    print('REF :', r['reference'])
    print('GEN :', r['prediction'])
    print('-'*60)
```
